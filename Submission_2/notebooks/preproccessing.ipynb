```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("movement_data.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
key_landmarks = {
    'Hombro Izq': 11,
    'Hombro Der': 12,
    'Codo Izq': 13,
    'Codo Der': 14,
    'Cadera Izq': 23,
    'Cadera Der': 24,
    'Rodilla Izq': 25,
    'Rodilla Der': 26,
    'Tobillo Izq': 27,
    'Tobillo Der': 28
}
```

```python
def extract_landmark_coords(df, idx):
    cols = [f"x_{idx}", f"y_{idx}", f"z_{idx}"]
    return df[cols].to_numpy(dtype=float)

def normalize_landmarks(df, key_landmarks):
    """Normaliza landmarks restando el centro de caderas y escalando por pelvis->tobillos."""
    lh = extract_landmark_coords(df, key_landmarks['Cadera Izq'])
    rh = extract_landmark_coords(df, key_landmarks['Cadera Der'])
    la = extract_landmark_coords(df, key_landmarks['Tobillo Izq'])
    ra = extract_landmark_coords(df, key_landmarks['Tobillo Der'])

    hip_mid = (lh + rh) / 2.0
    ankle_mid = (la + ra) / 2.0

    scales = np.linalg.norm(hip_mid - ankle_mid, axis=1)
    scales[scales == 0] = 1.0

    lm_indices = sorted({int(c.split("_")[1]) for c in df.columns if c.startswith("x_")})
    coords = {}
    for i in lm_indices:
        arr = extract_landmark_coords(df, i)
        arr_norm = (arr - hip_mid) / scales[:, None]
        coords[i] = arr_norm
    return coords, hip_mid, ankle_mid, scales
```

```python
coords_norm, hip_mid, ankle_mid, scales = normalize_landmarks(df, key_landmarks)
print(coords_norm)
```

```python
def smooth_coords(coords_dict, window=5):
    """Suavizado por media mvil centrada."""
    smoothed = {}
    for idx, arr in coords_dict.items():
        df_tmp = pd.DataFrame(arr, columns=["x","y","z"])
        df_s = df_tmp.rolling(window=window, center=True, min_periods=1).mean()
        smoothed[idx] = df_s.to_numpy()
    return smoothed
```

```python
coords_s = smooth_coords(coords_norm)
print(coords_s)
```

```python
def compute_velocities(smoothed_coords, fps=None):
    """Calcula velocidades vectoriales y magnitudes por landmark."""
    vel = {}
    for i, arr in smoothed_coords.items():
        d = np.vstack([np.zeros(3), np.diff(arr, axis=0)])
        if fps is not None:
            d = d * fps
        speed = np.linalg.norm(d, axis=1)
        vel[i] = {"vel_vec": d, "speed": speed}
    return vel
```

```python
velocities = compute_velocities(coords_s, fps=30)
print(velocities)
```

```python
def angle_between_vectors(a, b):
    dot = np.sum(a*b, axis=1)
    na = np.linalg.norm(a, axis=1)
    nb = np.linalg.norm(b, axis=1)
    cos = dot / (na * nb + 1e-8)
    cos = np.clip(cos, -1.0, 1.0)
    return np.arccos(cos)

def compute_joint_angles(smoothed_coords, key_landmarks):
    """ngulos principales (rodillas, caderas, hombros, codos)."""
    kl = key_landmarks

    # Extremidades inferiores
    hipL = smoothed_coords[kl['Cadera Izq']]
    hipR = smoothed_coords[kl['Cadera Der']]
    kneeL = smoothed_coords[kl['Rodilla Izq']]
    kneeR = smoothed_coords[kl['Rodilla Der']]
    ankleL = smoothed_coords[kl['Tobillo Izq']]
    ankleR = smoothed_coords[kl['Tobillo Der']]

    # Extremidades superiores
    shoulderL = smoothed_coords[kl['Hombro Izq']]
    shoulderR = smoothed_coords[kl['Hombro Der']]
    elbowL = smoothed_coords[kl['Codo Izq']]
    elbowR = smoothed_coords[kl['Codo Der']]

    # ngulos inferiores
    kneeL_ang = angle_between_vectors(hipL - kneeL, ankleL - kneeL)
    kneeR_ang = angle_between_vectors(hipR - kneeR, ankleR - kneeR)
    hipL_ang = angle_between_vectors(kneeL - hipL, ankleL - hipL)
    hipR_ang = angle_between_vectors(kneeR - hipR, ankleR - hipR)

    # ngulos superiores
    elbowL_ang = angle_between_vectors(shoulderL - elbowL, hipL - elbowL)
    elbowR_ang = angle_between_vectors(shoulderR - elbowR, hipR - elbowR)
    shoulderL_ang = angle_between_vectors(elbowL - shoulderL, hipL - shoulderL)
    shoulderR_ang = angle_between_vectors(elbowR - shoulderR, hipR - shoulderR)

    return {
        "rodilla_Izq": kneeL_ang,
        "rodilla_Der": kneeR_ang,
        "cadera_Izq": hipL_ang,
        "cadera_Der": hipR_ang,
        "codo_Izq": elbowL_ang,
        "codo_Der": elbowR_ang,
        "hombro_Izq": shoulderL_ang,
        "hombro_Der": shoulderR_ang
    }
```

```python
angles = compute_joint_angles(coords_s, key_landmarks)
print(angles)
```

```python
def trunk_inclination(smoothed_coords, key_landmarks):
    """Inclinacin del tronco (vector caderashombros vs vertical)."""
    hipL = smoothed_coords[key_landmarks['Cadera Izq']]
    hipR = smoothed_coords[key_landmarks['Cadera Der']]
    shoulderL = smoothed_coords[key_landmarks['Hombro Izq']]
    shoulderR = smoothed_coords[key_landmarks['Hombro Der']]

    hip_mid = (hipL + hipR) / 2.0
    shoulder_mid = (shoulderL + shoulderR) / 2.0
    vec = shoulder_mid - hip_mid
    vertical = np.tile(np.array([0.0, -1.0, 0.0]), (vec.shape[0], 1))
    ang = angle_between_vectors(vec, vertical)
    return ang
```

```python
trunk_ang = trunk_inclination(coords_s, key_landmarks)
print(trunk_ang)
```

```python
feature_parts = []
selected_indices = set(key_landmarks.values())
idx_to_name = {v: k.replace(" ", "_") for k, v in key_landmarks.items()}

# Solo landmarks definidos, con nombres legibles
for idx, arr in coords_s.items():
    if idx not in selected_indices:
        continue
    name = idx_to_name[idx]
    part = pd.DataFrame({
        f"x_{name}": arr[:, 0],
        f"y_{name}": arr[:, 1],
        f"z_{name}": arr[:, 2],
        f"velocidad_{name}": velocities[idx]["speed"]
    }, index=df.index)
    feature_parts.append(part)

# ngulos articulares
angle_df = pd.DataFrame(
    {f"{name}_ang": np.degrees(arr) for name, arr in angles.items()},
    index=df.index
)
feature_parts.append(angle_df)

# Inclinacin del tronco
feature_parts.append(
    pd.DataFrame({"inclinacion_tronco_ang": np.degrees(trunk_ang)}, index=df.index)
)

features = pd.concat(feature_parts, axis=1)

# Agregar etiquetas si existen
if "action" in df.columns:
    features["accion"] = df["action"].values
if "speed" in df.columns:
    features["velocidad_accion"] = df["speed"].values

features.columns = features.columns.str.lower()

features.head()

```

```python
features.to_csv("mov_data_proccesed.csv", index=False)
```
